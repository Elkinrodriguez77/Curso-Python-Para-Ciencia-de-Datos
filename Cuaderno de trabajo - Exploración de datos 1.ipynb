## Implementando operaciones en Pandas

Crear una columna de ganancias a partir de los datos cargados en la variable "Precios"

```python
Precios["Ganancias"] = Precios["Precio"] - Precios["Costo"]
```

¿Por qué arroja error el código anterior? Por el simbolo $, para esto debemos reemplazarlo:

```python
Precios["Costo"] = Precios["Costo"].str.replace("$","").astype("float")
```

Validar la nueva columna:

```python
Precios.head(10)
```

Eliminar la columna "Time" del dataframe de Ventas:

```python
del Ventas["Time"]
```

```python
Ventas.head(10)
```

Crear un nuevo dataframe con todos los productos con un precio mayor a 90.

```python
FiltroPrecios = Precios[(Precios["Precio"]>90) & (Precios["Categoria"]=="Mundo Tec")]
```

Validar el nuevo DataFrame

```python
FiltroPrecios.head(10)
```

Crear una columna que clasifique los productos en alto/bajo margen, si su ganancia es mayor a 100 pesos:

```python
Precios["Margen"] = np.where(Precios["Ganancias"]>100, 'Alto', 'Bajo')
```

## Buscarv / Join, Merge

Unir los 2 dataframes usando un left join:

```python
Buscarv_df = Ventas.merge(Precios, on=["Producto"], how="left")
```

```python
Buscarv_df.head(10)
```

## Tablas Dinámicas con Python

Dentro de las funcionalidades de Pandas, está la de analizar datos mediante tablas dinámicas

Crer una tabla dinámica con la tabla que unimos, que nos muestre las ventas por producto (De mayor a menor):

```python
Vet_x_Pro = pd.pivot_table(Buscarv_df, values="Precio", index="Producto", aggfunc=np.sum)
```
