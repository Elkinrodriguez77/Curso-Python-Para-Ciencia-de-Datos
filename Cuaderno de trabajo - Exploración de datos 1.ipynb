## Implementando operaciones en Pandas

Crear una columna de ganancias a partir de los datos cargados en la variable "Precios"

```python
Precios["Ganancias"] = Precios["Precio"] - Precios["Costo"]
```

驴Por qu茅 arroja error el c贸digo anterior? Por el simbolo $, para esto debemos reemplazarlo:

```python
Precios["Costo"] = Precios["Costo"].str.replace("$","").astype("float")
```

Validar la nueva columna:

```python
Precios.head(10)
```

Crear un nuevo dataframe con todos los productos con un precio mayor a 90.

```python
FiltroPrecios = Precios[(Precios["Precio"]>90) & (Precios["Categoria"]=="Mundo Tec")]
```

Validar el nuevo DataFrame

```python
FiltroPrecios.head(10)
```

Crear una columna que clasifique los productos en alto/bajo margen, si su ganancia es mayor a 100 pesos:

```python
Precios["Margen"] = np.where(Precios["Ganancias"]>100, 'Alto', 'Bajo')
```
